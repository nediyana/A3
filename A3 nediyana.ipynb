- General Findings

When I had everything besides 'running', my multi-class accuracies for the 3 classifiers (with just the means of acceleration on the Z axis as the features) were around 93%. 
However, after I added the running data, my accuracies dropped to around 70%. I tried a few things until I found that a combination of the accelerations on all 3 axes works best (but still not as high as without running).

However, acceleration on the Y axis is what classified 'walking' with the highest accuracy in all classifiers. 

Acceleration on the X axis performed the worst from all classifiers. 

- Which features were most useful?
I combined acceleration on all 3 axes: abs(accelerationX) + accelerationY + accelerationZ to get a value for each line, and then took the mean of all 100 lines within one 10-second chunk. That gave me pretty high accuracies: 87% for decision trees, 83% for SVM, and 80% for Logistic Regression.

- Which learning algorithms were most effective?
The most effective one was decision trees. 

- Which activity was hardest to classify?
The activity that was hardest to classify was running. As mentioned before, when I introduced it to my dataset, my accuracies went down, so I had to change the features array. 
Sitting and Walking were always classified correctly. However, running, climbing stairs, and riding in a car were miss-classified.  I think that's because I went up and down the stairs really quickly, so it almost looks like I'm running. As for being in car, Providence's roads are very bumpy, but we also stopped at many traffic lights, so I can see how the movement in the car might have resembled running/climbing stairs.
